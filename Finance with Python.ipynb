# Build a Simple Trading Strategy

## Munging the stock data and add two columns - MA10 and MA50

```python
fb['MA10'] = fb['Close'].rolling(10).mean()
fb['MA50'] = fb['Close'].rolling(50).mean()
fb = fb.dropna()
fb = fb.drop(columns=(['Dividends', 'Stock Splits', 'PriceDiff', 'Return', 'Direction', 'ma50']))
fb.head()
```

## Add "Shares" column to make decisions base on the strategy

```python
#Add a new column "Shares", if MA10>MA50, denote as 1 (long one share of stock), otherwise, denote as 0 (do nothing)

fb['Shares'] = [1 if fb.loc[ei, 'MA10']>fb.loc[ei, 'MA50'] else 0 for ei in fb.index]
```

```python
#Add a new column "Profit" using List Comprehension, for any rows in fb, if Shares=1, the profit is calculated
# as the close price of tomorrow - the close price of today. Otherwise the profit is 0.

#Plot a graph to show the Profit/Loss

fb['Close1'] = fb['Close'].shift(-1)
fb['Profit'] = [fb.loc[ei, 'Close1'] - fb.loc[ei, 'Close']\
                     if fb.loc[ei, 'Shares']==1 else 0 for ei in fb.index]
fb['Profit'].plot()
plt.axhline(y=0, color='red')
```

## Use .cumsum() to display our model's performance if we follow the strategy

```python
#Use .cumsum() to calculate the accumulated wealth over the period

fb['wealth'] = fb['Profit'].cumsum()
fb.tail()
```

```python
#plot the wealth to show the growth of profit over the period

fb['wealth'].plot()
plt.title('Total money you win is {}'.format(fb.loc[fb.index[-2], 'wealth']))
```

# Distribution of Log return

```python
# calculate the log daily return
fb['LogReturn'] = np.log(fb['Close']).shift(-1) - np.log(fb['Close'])
```

```python
# Plot a histogram to show the distribution of log return of Facebook's stock. 
# It is very close to a normal distribution
from scipy.stats import norm
mu = fb['LogReturn'].mean()
sigma = fb['LogReturn'].std(ddof=1)

density = pd.DataFrame()
density['x'] = np.arange(fb['LogReturn'].min()-0.01, fb['LogReturn'].max()+0.01, 0.001)
density['pdf'] = norm.pdf(density['x'], mu, sigma)

fb['LogReturn'].hist(bins=50, figsize=(15, 8))
plt.plot(density['x'], density['pdf'], color='red')
plt.show()
```

# Probability of the stock price will drop over a certain percentage in a day

```python
# probability that the stock price of facebook will drop over 5% in a day
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The Probability is ', prob_return1)
```

```python
# probability that the stock price of facebook will drop over 10% in a day
prob_return1 = norm.cdf(-0.10, mu, sigma)
print('The Probability is ', prob_return1)
```

# Probability of the stock price will drop over a certain percentage in a year

```python
# drop over 40% in 220 days
mu220 = 220*mu
sigma220 = (220**0.5) * sigma
print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))
```

```python
# drop over 20% in 220 days
mu220 = 220*mu
sigma220 = (220**0.5) * sigma
drop20 = norm.cdf(-0.2, mu220, sigma220)
print('The probability of dropping over 20% in 220 days is ', drop20)
```

# Value at risk (VaR)

```python
# Value at risk(VaR)
VaR = norm.ppf(0.05, mu, sigma)
print('Single day value at risk ', VaR)
```

```python
# Quatile 
# 5% quantile
print('5% quantile ', norm.ppf(0.05, mu, sigma))
# 95% quantile
print('95% quantile ', norm.ppf(0.95, mu, sigma))
```

# Hypothesis Testing

```python
# Log return goes up and down during the period
fb['LogReturn'].plot(figsize=(20, 8))
plt.axhline(0, color='red')
plt.show()
```

## Steps involved in testing a claim by hypothesis testing

### Calculate Test Statistic

```python
sample_mean = fb['LogReturn'].mean()
sample_std = fb['LogReturn'].std(ddof=1)
n = fb['LogReturn'].shape[0]

# if sample size n is large enough, we can use z-distribution, instead of t-distribtuion
# mu = 0 under the null hypothesis
z_dist = (sample_mean - 0)/(sample_std/n**0.5)
print(z_dist)
```

### Set Desicion Criteria

```python
# confidence level
alpha = 0.05

zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft  # z-distribution is symmetric 
print(zleft, zright)
```

### Make Decision - Shall We Reject H0?

```python
print('At significant level of {}, shall we reject: {}'.format(alpha, z_dist>zright or z_dist<zleft))
```

# Evaluating Strategy Built from Regression Model

```python
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")
```

```python
Train = fb.iloc[-2000:-1000, :]
Test = fb.iloc[-1000:, :]
```

```python
formula = 'Open~High+Low+Close+Volume+MA10+MA50+Shares+Close1+Profit+wealth'
lm = smf.ols(formula=formula, data=Train).fit()
```

```python
Train['PredictedY'] = lm.predict(Train)
Test['PredictedY'] = lm.predict(Test)
```

## Profit of Signal-Based Strategy

```python
# Train
Train['Order'] = [1 if sig>0 else -1 for sig in Train['PredictedY']]
Train['Profit'] = Train['Open'] * Train['Order']

Train['Wealth'] = Train['Profit'].cumsum()
print('Total profit made in Train: ', Train['Profit'].sum())
```

```python
plt.figure(figsize=(10, 10))
plt.title('Performance of Strategy in Train')
plt.plot(Train['Profit'].values, color='green', label='Signal based strategy')
plt.plot(Train['Open'].cumsum().values, color='red', label='Buy and Hold strategy')
plt.legend()
plt.show()
```

```python
# Test
Test['Order'] = [1 if sig>0 else -1 for sig in Test['PredictedY']]
Test['Profit'] = Test['Open'] * Test['Order']

Test['Wealth'] = Test['Profit'].cumsum()
print('Total profit made in Test: ', Test['Profit'].sum())
```

```python
plt.figure(figsize=(10, 10))
plt.title('Performance of Strategy in Train')
plt.plot(Test['Wealth'].values, color='green', label='Signal based strategy')
plt.plot(Test['Open'].cumsum().values, color='red', label='Buy and Hold strategy')
plt.legend()
plt.show()
```
